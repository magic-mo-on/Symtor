```python
import tabula
df = tabula.read_pdf("vaccination.pdf", pages='all')
```

```python
main_dict={"vaccinationCenter":list(),"address":list(),"state":list(),"district":list()}
```

```python
for data in range(1,len(df)):
    cols=df[data].columns
    main_dict["vaccinationCenter"].append(cols[0])
    main_dict["address"].append(cols[1])
    main_dict["state"].append(cols[2])
    main_dict["district"].append(cols[3])
    for index, row in df[data].iterrows():
        main_dict["vaccinationCenter"].append(row[cols[0]])
        main_dict["address"].append(row[cols[1]])
        main_dict["state"].append(row[cols[2]])
        main_dict["district"].append(row[cols[3]])
```
