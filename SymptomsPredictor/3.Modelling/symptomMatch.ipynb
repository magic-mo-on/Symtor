```python
symptom_occurance_dict=dict()
for disease,symptoms in diseases_symptoms_cleaned.items():
    for symptom in symptoms:
        if symptom in symptom_occurance_dict:
            symptom_occurance_dict[symptom]+=1
        else:
            symptom_occurance_dict[symptom]=1
```

```python
import operator
symptom_occurance_list=list()
symptom_occurance_dict = dict(sorted(symptom_occurance_dict.items(), key=operator.itemgetter(1),reverse=True))
# print('Dictionary in descending order by value : ',symptom_occurance_dict)
for disease,occurance in symptom_occurance_dict.items():
    #print(disease,"-->",occurance)
    symptom_occurance_list.append(disease)
print(symptom_occurance_list[:20])
```

# Creating dictonary of symptom and diseases that has that symptom

```python
symptom_disease_dict=dict()
for disease,symptoms in diseases_symptoms_cleaned.items():
    for symptom in symptoms:
        if symptom in symptom_disease_dict:
            symptom_disease_dict[symptom].append(disease)
        else:
            symptom_disease_dict[symptom]=[disease]
```
